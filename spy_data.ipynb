```python
# Imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import alpaca_trade_api as tradeapi
%matplotlib inline
```

```python
from dotenv import load_dotenv
load_dotenv()

# Set Alpaca API key and secret Key
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2"
)
```

```python
ticker = "SPY"

# Set timeframe to '1D'
timeframe = "1D"

start_date = pd.Timestamp("2020-01-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2020-12-31", tz="America/New_York").isoformat()

spy_data = api.get_barset(
    ticker,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,
).df
spy_data.head()
```

```python
spy_close = pd.DataFrame(spy_data["SPY"]["close"])
spy_close.index = spy_close.index.date
spy_close.head()
```
